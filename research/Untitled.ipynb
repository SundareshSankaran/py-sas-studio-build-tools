```python
cs = []

for step in os.listdir(data_folder):
    if step!= ".DS_Store":
        cs.append(CustomStep(name=os.path.splitext(os.path.basename(step))[0]).load_step_file(custom_step_file=os.path.join(data_folder,step)))
```

```python
from deepdiff import DeepDiff
import pandas as pd

dict1 = cs[0]
dict2 = cs[1]

diff = DeepDiff(dict1, dict2, view='tree')

rows = []
for change in diff.get('values_changed', []):
    rows.append({
        'Key': change.path(),
        'dict1': change.t1,
        'dict2': change.t2,
        'Change': 'value_changed'
    })
for change in diff.get('type_changes', []):
    rows.append({
        'Key': change.path(),
        'dict1': f"{change.t1} ({type(change.t1).__name__})",
        'dict2': f"{change.t2} ({type(change.t2).__name__})",
        'Change': 'type_changed'
    })
for change in diff.get('dictionary_item_added', []):
    rows.append({
        'Key': change.path(),
        'dict1': '',
        'dict2': change.t2,
        'Change': 'added'
    })
for change in diff.get('dictionary_item_removed', []):
    rows.append({
        'Key': change.path(),
        'dict1': change.t1,
        'dict2': '',
        'Change': 'removed'
    })

df = pd.DataFrame(rows, columns=['Key', 'dict1', 'dict2', 'Change'])
print(df.to_markdown(index=False))
```
